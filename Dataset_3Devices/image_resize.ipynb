```python
target_image_size = (75, 75)
```

```python
import os
from PIL import Image

def resize_images(folder_path):
    # 遍歷資料夾中的所有檔案和子資料夾
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            # 判斷是否為圖片檔案
            if file.endswith(".jpg") or file.endswith(".jpeg") or file.endswith(".png"):
                # 讀取圖片
                img_path = os.path.join(root, file)
                img = Image.open(img_path)

                # 將圖片 resize 成 target_image_size
                img = img.resize(target_image_size)

                # 儲存 resized 圖片
                resized_img_path = os.path.join(root, file)
                img.save(resized_img_path)

                print(f"{img_path} resized and saved as {resized_img_path}")

# 指定資料夾路徑
folder_path = "./Nokia/train"

# 呼叫 resize_images 函數
resize_images(folder_path)
```
